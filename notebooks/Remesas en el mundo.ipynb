# Remesas en el mundo
En este notebook haremos un análisis de las remesas del mundo con datos de World Bank.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rem_df = pd.read_csv("../data/API_BX/remittances.csv", skiprows = 4)
rem_df
```

```python
df = rem_df[rem_df["Country Code"].isin(["MEX", "CHN", "IND", "PHL", "FRA", "DEU"])]
```

```python
df.iloc[:, 2:26]
```

```python
df.drop(df.iloc[:, 2:26], inplace = True, axis = 1)
```

```python
df
```

```python
df_melt = pd.melt(df, id_vars = ["Country Name", "Country Code"])
```

```python
df_melt["variable"].unique()
```

```python
df_melt[df_melt.variable == "Unnamed: 65"]
```

```python
df_melt = df_melt[df_melt["variable"] != "Unnamed: 65"]
df_melt
```

```python
from plotnine import *
flip_xlabels = theme(axis_text_x = element_text(angle=90, hjust=1))
```

```python
df_melt["value"] = df_melt["value"]/1000000 # Para mostrar los valores en millones
```

```python
(ggplot(df_melt, aes(x = "variable", y = "value", color = "Country Code", group = "Country Code"))+
    geom_line()+
    labs(x = "Año", y = "Remesas (millones de USD)", color = "País")+
    theme_classic()+
    flip_xlabels)
```

## ¿Quiénes son los países con mayor recepción de remesas a 2020?

```python
rem_df
```

```python
rem_df.drop(rem_df.iloc[:, 2:26], inplace = True, axis = 1) # Reducir los años arriba de 1981
```

```python
rm_df = pd.melt(rem_df, id_vars = ["Country Name", "Country Code"])
```

```python
sorted_df = rm_df[rm_df["variable"] == "2020"].sort_values(by = ["value"], ascending = False)
```

```python
df2 =sorted_df[~sorted_df["Country Code"].isin(["WLD", "IBT", "LMY", "MIC", "IBD", "EAR",
                                          "LMC", "OED", "ECS", "UMC", "SAS", "TSA",
                                          "PST", "HIC", "LTE", "IDA", "EUU", "EAS", "LCN",
                                          "TLA", "LAC", "TEA", "EAP", "EMU","TEC","IDX",
                                          "ECA", "MEA", "ARB", "MNA", "IDB", "TMN", "LDC",
                                          "FCS", "TSS", "SSF", "SSA", "PRE", "CEB", "HPC",
                                          "AFW"])]
df2
```

```python
# Gráfica por ingresos
sorted_df[sorted_df["Country Code"].isin(["MIC", "LMC", "UMC", "HIC"])]
```

```python
df2["value"] = df2["value"]/1000000
```

```python
(ggplot(df2.iloc[0:10], aes(x = "Country Name", y = "value"))+
     geom_bar(stat = "identity")+
     labs(x = "País", y = "Remesas (millones de USD)")+
     ggtitle("2020")+
     theme_classic()+
     theme(axis_text_x = element_text(angle=60, hjust=1)))
```
