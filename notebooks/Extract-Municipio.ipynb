```python
import pandas as pd

df = pd.read_csv("../data/remesas-mx.csv", encoding="latin")
```

```python
df = df.drop([0,1])
df.head()
```

```python
municipios = [name[:48] == "Ingresos por Remesas, Distribución por Municipio" for name in df.columns]
```

```python
df_municipios = df.iloc[:,municipios]
```

```python
df_municipios.head()
```

```python
df_municipios['Fecha'] = df['Título']
```

```python
df_mun = pd.melt(df_municipios, id_vars="Fecha")
df_mun["Estado"] = [estado.split(",")[2][1:] for estado in df_mun["variable"]]
df_mun["Municipio"] = [mun.split(",")[3][1:] for mun in df_mun["variable"]]
df_mun = df_mun.drop(columns="variable")
```

```python
remesas_dgo = df_mun[df_mun["Estado"] == "Durango"]
```

```python
df_mun.to_csv("remesas_mun_mx.csv")
```
